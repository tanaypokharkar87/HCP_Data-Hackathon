```python
def getchannel(matrix):
    ans = "Sales_Rep_Calls"

    if matrix['Brand_Rx'][3]>matrix['Brand_Rx'][2]:
        if matrix['Brand_Rx'][3]>matrix['Brand_Rx'][4]:
            ans = "Emails_Delivered";
        else :
            ans = "Samples_Dropped"
    else :
        if matrix['Brand_Rx'][2]<matrix['Brand_Rx'][4]:
            ans = "Samples_Dropped"
    return ans;


    
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn import preprocessing

def randomForestModel(X_train,y_train,X_test):

#     rfc = RandomForestRegressor(n_estimators = 10, random_state = 0)

#     rfc.fit(X_train, y_train)

#     y_pred = rfc.predict(X_test)

#     return y_pred;
    rfc = RandomForestRegressor(n_estimators = 10, random_state = 0)

    rfc.fit(X_train, y_train)

    parameters = {
        'n_estimators': [10],
        'max_depth': [1],
    }

    clf = GridSearchCV(rfc, parameters)
    
    clf.fit(X_train, y_train)
    
    y_pred = clf.predict(X_test)

    return y_pred;


# Check accuracy score 

# from sklearn.metrics import accuracy_score
# print('Model accuracy '. format(accuracy_score(y_test, y_pred)))
```

```python
def findChannels(a,b):
    data4 = data3.iloc[a:b]
    X=data4.drop(['Physician_ID_0'],axis=1)
    X=X.drop(['Physician_ID_1'],axis=1)
    X=X.drop(['Physician_ID_2'],axis=1)
    X=X.drop(['Physician_ID_3'],axis=1)
    X=X.drop(['Physician_ID_4'],axis=1)
    X=X.drop(['Physician_ID_5'],axis=1)
    X=X.drop(['Physician_ID_6'],axis=1)
    X=X.drop(['Physician_ID_7'],axis=1)
    X=X.drop(['Physician_ID_8'],axis=1)
    X=X.drop(['Physician_ID_9'],axis=1)
    X=X.drop(['Physician_ID_10'],axis=1)
    X=X.drop(['Physician_ID_11'],axis=1)
    X=X.drop(['Physician_ID_12'],axis=1)
    X=X.drop(['Physician_ID_13'],axis=1)
    X=X.drop(['Time_Period'],axis=1)
    X=X.drop(['Brand_Rx'],axis=1)
    X=X.drop(['Physician_Segment'],axis=1)
    X=X.drop(['Specialty'],axis=1)
    X=X.drop(['Speaker_Programs_Attended'],axis=1)
    X=X.drop(['Vouchers_Dropped'],axis=1)
#     X=X.drop(['Unnamed: 0'],axis=1)
    y=data4['Brand_Rx']
    
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.333, random_state =42 )
    
    encoder = ce.OrdinalEncoder(cols=['Sales_Rep_Calls', 'Samples_Dropped', 'Emails_Delivered'])


    X_train = encoder.fit_transform(X_train)

    X_test = encoder.transform(X_test)
    
    y_pred = randomForestModel(X_train,y_train,X_test);
    
    df = {
        'Brand_Rx':y_pred,
        'Market_Rx' : X_test['Market_Rx'],
        'Sales_Rep_Calls' : X_test['Sales_Rep_Calls'],
        'Samples_Dropped' :X_test['Samples_Dropped'],
        'Emails_Delivered' : X_test['Emails_Delivered']
    }
    
#     print("hello1")
#     print(df)
    
    dfa = pd.DataFrame(df)
#     print(dfa)
#     print("hello2")
    matrix = dfa.corr()
    
    channel = getchannel(matrix)
    
    return channel
    
```

```python
for i in range(10000):
    channel = findChannels(a,b)
    print(i," ",channel," ",a," ",b)
    
    bestChannel.append(channel)
    a += 57
    b += 57
```
