```python
import pandas as pd
```

```python
data1 = pd.read_excel('/kaggle/input/kdsh-data-solution-task-2/HCP_Data_KDAG_Hackathon/HCP_Data_KDAG_Hackathon.xlsx')
```

```python
data1.head()
```

```python
data = {
    "Time Period":data1['Time_Period'],
    "Brand_Rx":data1['Brand_Rx']
}
```

```python
df= pd.DataFrame(data)
```

```python
x = df.iloc[1:57]
```

```python
months= []
years=[]
day=[]
import numpy as np
sliced_text = slice(6)
for i in x['Time Period']:
    years.append((int)(str(i)[0:4]));
    months.append((int)(str(i)[4:6]));
    day.append((int)(str(i)[6:8]));

# convert into one matrix
X = np.array([day, months, years]).T
```

```python
def findBrandRx(a,b):
    x1 = df.iloc[a:b]
    from sklearn.ensemble import RandomForestRegressor

    # fit the model
    my_rf = RandomForestRegressor()
    my_rf.fit(X, x1['Brand_Rx'])

    # predict for the feb 1st week
    day1 = [7]
    month1 = [2]
    year1 = [2020]

    X1 = np.array([day1,month1,year1]).T

    preds = my_rf.predict(X1)
    
    return preds;
```

```python
a=1
b=57
Estimated_Rx = []
for i in range(10000):
    brand_Rx=findBrandRx(a,b)
    Estimated_Rx.append(brand_Rx)
    print(a," ",b," ",brand_Rx)
    a += 57
    b += 57
```

```python
Estimated_Rx
```

```python
Expected_Trx = []
for i in Estimated_Rx:
    Expected_Trx.append(i[0])
```
